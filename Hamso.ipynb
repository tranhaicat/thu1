
# **Định nghĩa hàm số**
Để hiểu được hàm số là gì, hãy xét hai dãy số $x$ và $y$ dưới đây

```python
x = [0,1,2,3,4,5] # giay
y = [0,1.5,3.6,9.3,16.3,23.5] # met
print('x =',x)
print('y =',y)
```

Chúng có số lượng phần tử như nhau. Phần tử đầu tiên của $y$ tương ứng với phần tử đầu tiên của $x$, phần tử thứ $i$ của $y$ tương ứng với phần tử thứ $i$ của $x$. Dãy $x$ ta gọi là biến số, còn dãy $y$ ta gọi là đối số. Hai dãy này kẹp vào nhau, cứ mỗi $x$ tồn tại đều được đáp trả với mỗi $y$ tương ứng như vậy, ta gọi là hàm số, kí hiệu $y(x)$.

Đồ thị của hàm số $y(x)$ là tập hợp các điểm có tọa độ $(x_i,y_i)$ trên hệ trục tọa độ. Với hệ trục tọa độ Decasters, đồ thị của hàm số trên biểu diễn thành bộ các điểm như hình dưới.

```python
plt.plot(x,y,'ro')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.show()
```

Nhiều khi  ta nối các điểm lại với nhau bằng đoạn thẳng:

```python
plt.plot(x,y,'-rs')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.show()
```


# **Hàm số giải tích**
Trong toán học, hàm số giải tích có định nghĩa chặt chẽ hơn. Trong bài viết này ta chỉ để ý đến một tính chất quan trọng của hàm giải tích: giữa hai điểm $(x_i,y_i)$ và $(x_j,y_j)$ luôn tồn tại (hay tìm thấy) một điểm $(x_k,y_k)$ chen giữa. 

Nói cách khác, hàm giải tích có vô số phần tử. Khi “phóng đại” một góc nhỏ trên đồ thị của một hàm giải tích, đồ thị không bị giãn ra thành các điểm rời rạc xa nhau mà luôn luôn có các điểm trung gian lấp đầy.

Vì vậy, ta không thể diễn tả một hàm giải thích dưới dạng bảng, dạng hai dãy số rời rạc, mà thông qua biểu thức đại số. 

Ví dụ xét hàm số giải tích $y=x^2+x+1$.  Hàm này có thể định nghĩa bằng ngôn ngữ Python như sau:

```python
def parabol_1(x):
    return x**2+x+1
```

Định nghĩa này cho phép ta diễn giải hàm số dưới dạng hai dãy số. Biến số $x$ được gán cụ thể, ví dụ từ -3 đến 3 với 100 điểm cách đều nhau:

```python
x = np.linspace(-3,3,100)
x
```

Từ đây đối số $y$ sẽ có đúng 100 phần tử miêu tả qua biểu thức  $y=x^2+x+1$:

```python
y = parabol_1(x)
y
```

Đồ thị hàm số với 100 điểm $(x_i,y_i)$ này đủ dày, khiến ta cảm tưởng về một đường cong mịn của một hàm số giải tích:

```python
plt.plot(x,y,'g-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# **Hàm số cho dưới dạng tham số**

Xét một chuyển động có hoành độ $x$ và tung độ $y$ thay đổi theo thời gian với quy luật:
$$x=3\cos(\omega t)$$
$$y=1.5\sin(\omega t)$$
với $\omega = 1$

```python
omega = 1
t = np.linspace(0,10,100)
x = np.cos(omega*t)
y = np.sin(omega*t)
```

Thực chất, $x(t)$ đã là một hàm số của tọa độ $x$ theo thời gian $t$, có nghĩa chuỗi thời gian $t$ trôi đi sẽ tương ứng với hoành độ $x$ biến thiên theo chuyển động của vật. Điều tương tự xảy ra với $y(t)$.  Kết quả là, cứ mỗi hoành độ $x$ cũng sẽ tương ứng với một tung độ $y$. Nói cách khác, hai dãy giá trị $x$ và $y$ cũng có thể kẹp thành một hàm số, gọi $y(x)$.

Có thể vẽ trực tiếp sự phụ thuộc của $y$ theo $x$, hay đồ thị hàm số $y(x)$.

Trong vật lý, hàm $y(x)$ gọi là phương trình quỹ đạo, còn đồ thị của nó thể hiện hình dáng của quỹ đạo.

```python
plt.plot(x,y,'-g')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.show()
```

Phương trình quỹ đạo trên có dạng elip.

Đôi khi ta cũng đặt câu hỏi về tiến trình thay đổi vị trí của vật theo thời gian, lúc ấy ta lại cần đồ thị của tọa độ $x$ và $y$ phụ thuộc vào thời gian, hay đồ thị hàm số $x(t)$ và $y(t)$, tức đồ thị của phương trình chuyển động:

```python
plt.subplot(2,1,1)
plt.plot(t,x,'-b')
plt.xlabel('t [s]')
plt.ylabel('x [m]')

plt.subplot(2,1,2)
plt.plot(t,y,'-r')
plt.xlabel('t [s]')
plt.ylabel('y [m]')
plt.show()
```
